```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/content/LabourTrainingEvaluationData.csv")
df.head()

# Assuming your DataFrame is named 'df' and you want to regress 'earnings' on 'training'
# Replace 'earnings' and 'training' with your actual column names.

# Create the OLS model
model = sm.ols('Earnings_1974 ~ Age', data=df).fit()

# Print the regression results
print(model.summary())

# 1. Scatter plot of Earnings_1974 vs. Age
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Earnings_1974', data=df)
plt.title('Scatter Plot of Earnings vs. Age')
plt.xlabel('Age')
plt.ylabel('Earnings in 1974')
plt.show()

# 2. Regression line plot
plt.figure(figsize=(8, 6))
sns.regplot(x='Age', y='Earnings_1974', data=df)
plt.title('Regression Plot of Earnings vs. Age')
plt.xlabel('Age')
plt.ylabel('Earnings in 1974')
plt.show()

# 3. Distribution of Residuals
residuals = model.resid
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()

# 4. Residuals vs. Fitted values plot
fitted_values = model.fittedvalues
plt.figure(figsize=(8, 6))
plt.scatter(fitted_values, residuals)
plt.axhline(y=0, color='r', linestyle='-')  # Add a horizontal line at y=0
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```
